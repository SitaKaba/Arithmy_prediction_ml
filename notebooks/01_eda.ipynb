# 01 – EDA (UCI Heart Disease – Cleveland)

```python

import pandas as pd
import matplotlib.pyplot as plt
from src.data import load_heart, get_feature_types

df = load_heart()
num, cat = get_feature_types()
df.head()
```

```python

df.describe(include='all').T
```

```python

# Target balance
df['target'].value_counts(normalize=True).plot(kind='bar')
plt.title('Target balance')
plt.show()
```

```python

# Simple numeric histograms
for col in num:
    df[col].hist(bins=20)
    plt.title(col)
    plt.xlabel(col); plt.ylabel('count')
    plt.show()
```
